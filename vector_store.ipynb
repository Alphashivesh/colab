```python
from langchain_community.vectorstores import Chroma
from langchain_ollama.embeddings import OllamaEmbeddings
from langchain_core.documents import Document
```

```python
Documents = [
    Document(page_content="Langchain is an open-source platform that enables us to build llm app."),
    Document(page_content="Open-ai is an closed-source model that provides their api to run your llm."),
    Document(page_content="Huggingface is one of the open-source platform for different models."),
    Document(page_content="Ollama is also one of the open-source platform for models but it provides a lot of flexibility."),
]
```

```python
embedding_model = OllamaEmbeddings(model ="llama3.2:3b")
```

```python
vectorstore = Chroma.from_documents(
    documents=Documents,
    embedding=embedding_model,
    collection_name = "my_collection"
)
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k":2})
```

```python
query = "What is chroma used for?"
result = retriever.invoke(query)
```

```python
for i, doc in enumerate(result):
    print(f"{i+1}. {doc}")
```
